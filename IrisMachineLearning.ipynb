```python
# Imports

from keras.models import *
from keras.layers import *
from keras import Input
from keras import Sequential
from tensorflow import keras

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

from sklearn.model_selection import train_test_split

import numpy as np
from keras import metrics
from sklearn.preprocessing import OneHotEncoder, StandardScaler

```

```python
path = "irisData.txt"
```

```python
#Load Dataset

data = pd.read_csv(path)
```

```python
data
```

```python
# Scale all values

X = data.iloc[ : , 0:4 ]

maxNums = X.max()

X['x1'] = X['x1']/maxNums[0]
X['x2'] = X['x2']/maxNums[1]
X['x2'] = X['x3']/maxNums[2]
X['x3'] = X['x4']/maxNums[3]


X
```

```python
# Hot encode the Y values

Y = data.iloc[ : , 4:5 ]

enc = OneHotEncoder()

Y = enc.fit_transform(Y).toarray()

Y = pd.DataFrame(Y)

x_train , x_test , y_train , y_test = train_test_split(X,Y, test_size=0.2, random_state=123)
```

```python
X
```

```python
Y
```

```python
model = Sequential()

model.add( Input(shape=(4,), dtype="float32"   )  )
model.add( Dense(3, activation='sigmoid')  )
model.add( Dense(3, activation='relu')  )
model.add( Dense(3, activation='sigmoid')  )

model.summary()
```

```python
#compile model: add optimizer and Loss

model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=['accuracy' , 'Recall' , 'Precision'])
```

```python
model.fit( x_train, y_train, epochs=10, batch_size=128, validation_split=0.1)
```

```python
model.get_weights()
```

```python
model.evaluate(x_test,y_test)
```

```python
def get_flower_type(arg1, arg2 , arg3, arg4):
    modelIris = keras.models.load_model('ps1modelIris')

    preVec = [arg1/maxNums[0], arg2/maxNums[1] , arg3/maxNums[2], arg4/maxNums[3]] 
    rVec = np.expand_dims( preVec,0)
        
    result = modelIris.predict(rVec)
        
    tClass = result.max()
    
    if(tClass == result[0][0]):
        
        print('Iris-setosa')
        
    if(tClass == result[0][1]):
        
        print('Iris-versicolor')    
        
    if(tClass == result[0][2]):
        
        print('Iris-virginica')
        
    print(result)
```
